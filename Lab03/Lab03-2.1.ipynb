```python
import random
import math
```

```python
def estimate_pi_until_stop(convergence, sentinel):
    inside = 0
    draws = 0
    while True:
        draws += 1
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x*x + y*y <= 1:
            inside += 1
        pi_estimate = 4 * inside / draws
        percent_diff = abs(pi_estimate - math.pi) / math.pi
        if percent_diff < convergence:
            stop_reason = "converged"
            break
        if draws >= sentinel:
            stop_reason = "draw limit exceeded"
            break
    return pi_estimate, draws, percent_diff, stop_reason
```

```python
convergence = float(input("Enter convergence criterion (e.g. 0.001): "))
sentinel = int(input("Enter maximum number of draws (sentinel) (e.g. 100000): "))

print("You entered:")
print("convergence =", convergence)
print("sentinel =", sentinel)
```

```python
pi_est, draws_used, pct_diff, reason = estimate_pi_until_stop(convergence, sentinel)

print("\n--- Result ---")
print("Stop reason:", reason)
print("Estimated pi:", pi_est)
print("True pi     :", math.pi)
print("Draws used  :", draws_used)
print("Percent diff:", pct_diff)
```
