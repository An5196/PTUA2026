```python
def rotate_letter(c, n):
    if not c.isalpha():
        return c

    if c.islower():
        start = ord('a')
    else:
        start = ord('A')

    return chr((ord(c) - start + n) % 26 + start)
```

```python
def rotate_word(s, n):
    result = ''
    for c in s:
        result += rotate_letter(c, n)
    return result
```

```python
rotate_word("cheer", 7)
```

```python
rotate_word("melon", -10)
```

```python
rotate_word("IBM", -1)
```
